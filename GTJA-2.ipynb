###  Method II - Pearson 

```python
import numpy as np
from scipy.stats import pearsonr

# Significance level
significance_level = 0.05

# Initialize dictionaries to store correlations and p-values
Correlations = {}
P_values = {}

# Variable name of the target column
y_var = 'y_value'  # Ensure this is the correct column name

# Get the values of the target column
y_data = df_Cleaned[y_var].values
```

```python
# Loop through each column in the dataframe
Res= []
for i in df_Cleaned.columns:
    if i != y_var:
        x_data = df_Cleaned[i].values
        Correlations[i], P_values[i] = pearsonr(y_data, x_data)
        
        

        # Check significance level and print if below threshold
        if P_values[i] < 0.05: #and Correlations[i]>0.5 : 
            #print(i, "   CORR:", Correlations[i],"    P_VALUE:", P_values[i])
            Res.append([i, Correlations[i],  P_values[i]]) 

correlations_df = pd.DataFrame(Res, columns=['Variable', 'Correlation', 'P_value'])

# # Add ranking based on correlation
# correlations_df['Correlation_Rank'] = abs(correlations_df['Correlation']).rank(ascending=False)
            
# print(correlations_df)        


# Add ranking based on absolute correlation values
correlations_df['Correlation_Rank'] = abs(correlations_df['Correlation']).rank(ascending=False)

# Sort the DataFrame by the absolute value of correlations in descending order
correlations_df = correlations_df.sort_values(by='Correlation_Rank')

# Print the DataFrame
print(correlations_df)
```
